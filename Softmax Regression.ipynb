```python
import glob
import matplotlib.pyplot as plt
from PIL import Image

import torch
```

### Loss

```python
def cal_dist(x1, y1, x2, y2, a, b):
    area = abs((x1-a) * (y2-b) - (y1-b) * (x2 - a))
    AB = ((x1-x2)**2 + (y1-y2)**2)
    distance = area/AB
    return distance
```

```python
x1 = torch.arange(0, 10)
y1 = torch.arange(0, 20, 2)

b = 0
val = 0
for x1_, y1_ in zip(x1, y1):
    val += cal_dist(x1[0], x1[0]*2+b, x1[1], x1[1]*2+b, x1_, y1_)

plt.title("loss : {}".format(val))
plt.scatter(x1, y1)
plt.plot(x1, x1*2+b, c='r')
plt.xlabel('time')
plt.ylabel('score')
plt.xlim([0, 10])
plt.ylim([0, 20])
plt.savefig('result/loss5')
plt.show()
```
