```python
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
```

### Regression 이란?

- 여러 개의 독립변수와 한 개의 종속변수 간의 상관관계를 모델링하는 기법

```python
x1 = torch.arange(0, 4)
y1 = torch.arange(0, 8, 2)

x2 = torch.arange(8, 16)
y2 = torch.arange(16, 32, 2)

plt.scatter(x1, y1, c='b')
plt.scatter(x2, y2, c='b')
plt.xlabel('time')
plt.ylabel('score')
plt.savefig('reg_1.png')
plt.show()
```

```python
k = torch.arange(0, 16)
reg = k*2

x1 = torch.arange(0, 4)
y1 = torch.arange(0, 8, 2)

x2 = torch.arange(8, 16)
y2 = torch.arange(16, 32, 2)

plt.scatter(x1, y1, c='b')
plt.scatter(x2, y2, c='b')
plt.plot(k, reg, c='r', linestyle='--')
plt.xlabel('time')
plt.ylabel('score')
plt.savefig('reg_2.png')
plt.show()
```

#### Regression to Classification

```python
k = torch.arange(0, 16)
reg = k*2
decision = k*0+10

x1 = torch.arange(0, 4)
y1 = torch.arange(0, 8, 2)

x2 = torch.arange(8, 16)
y2 = torch.arange(16, 32, 2)

plt.scatter(x1, y1, c='b')
plt.scatter(x2, y2, c='b')
plt.plot(k, decision, c='y')
plt.plot(k, reg, c='r', linestyle='--')
plt.savefig('reg_class_1.png')
plt.xlabel('time')
plt.ylabel('score')
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

def linear(x, a=1/20, b=0.5):
    return a*x+b

k = torch.arange(-10, 10, 0.1)
```

```python
x1x2 = []

for i in range(-9, 10):
    for j in range(1, 10):
        x1x2.append([i, j/10])
        
x1x2 = torch.FloatTensor(x1x2)
```

### (x1, x2) Visualization

```python

for x_1, x_2 in x1x2:
    if sigmoid(x_1) <= x_2:
        plt.scatter(x_1, x_2, c='black')
    else :
        plt.scatter(x_1, x_2, c='black')
        

plt.title('Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.xlim([-10, 10])
plt.ylim([0, 1])
plt.savefig('result/data1.png')
plt.show()
```

### (x1, x2, y) Visualization

```python
for x_1, x_2 in x1x2:
    if sigmoid(x_1) <= x_2:
        plt.scatter(x_1, x_2, c='r')
    else :
        plt.scatter(x_1, x_2, c='b')

plt.title('Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.xlim([-10, 10])
plt.ylim([0, 1])
plt.savefig('result/data2.png')
plt.show()
```

### Linear Regression (change a)

```python
for x_1, x_2 in x1x2:
    if sigmoid(x_1) <= x_2:
        plt.scatter(x_1, x_2, c='r')
    else :
        plt.scatter(x_1, x_2, c='b')
        
plt.plot(k, linear(k, a=1/40, b=0.5), c='green')

plt.title('Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.xlim([-10, 10])
plt.ylim([0, 1])
plt.savefig('result/linear4.png')
plt.show()
```

### Linear Regression (change b)

```python
for x_1, x_2 in x1x2:
    if sigmoid(x_1) <= x_2:
        plt.scatter(x_1, x_2, c='r')
    else :
        plt.scatter(x_1, x_2, c='b')
        
plt.plot(k, linear(k, a=1/20, b=0.3), c='green')

plt.title('Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.xlim([-10, 10])
plt.ylim([0, 1])
plt.savefig('result/linear_b_2.png')
plt.show()
```

### Logistic Regression (change b)

```python
for x_1, x_2 in x1x2:
    if sigmoid(x_1) <= x_2:
        plt.scatter(x_1, x_2, c='r')
    else :
        plt.scatter(x_1, x_2, c='b')
        
plt.plot(k, sigmoid(k), c='green')

plt.title('Classification')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.xlim([-10, 10])
plt.ylim([0, 1])
plt.savefig('result/linear_b_2.png')
plt.show()
```
